```python
import csv
import os
import cv2
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
df = pd.read_csv('result.csv')
df
```

```python
folder_path = 'train_image/train_image'
cnt = 0
for row in df.itertuples():
    file_name = row[2][:-3]+'txt'
    class_id = 0
    center = row[4]
    center = center.strip('()').split(',')
    center_x = float(center[0])/512
    center_y = float(center[1])/512
    width = row[-1]/512
    height = row[-2]/512
    with open(f'{folder_path}/{file_name}', 'w') as f:
        f.write(f"{class_id} {center_x} {center_y} {width} {height}")
    cnt += 1
    print(f'{cnt}/{len(df)}')
```
